```python
# https://programmers.co.kr/learn/courses/30/lessons/12952
# 대표적인 풀이

def dfs(queen, row, n):
    count = 0
    if n == row: # 1 n==row 모든 행에 퀸을 배치했다. -> 종료
        return 1
    for col in range(n): # 2 
        queen[row] = col
        for i in range(row): # 3
            if queen[i] == queen[row]: #4 값이 같다 = 같은 열에 있다
                break
            if abs(queen[i]-queen[row]) == row - i: #5 abs(절대값)으로 대각선 오른쪽 왼쪽에 퀸이 있는지 파악
                break
        else: # 6 for-else 구문 사용 for-else 사이에서 break가 일어나지 않으면 else가 실행됨
            count += dfs(queen, row + 1, n)
    return count

def solution(n):
    return dfs([0]*n, 0, n)
```

```python
# https://sung-jun.tistory.com/m/112

# 열, 행, 대각선 경우 다 포함
n=int(input())
pos = [0] * n  # 각 열에서 퀸의 위치를 출력, pos[index] = x 일 때, index는 현재 열, x는 행
flag = [False] * n  # 각 행에 퀸을 배치했는지 체크
flag_a = [False] * (n*2-1)   # ↗  대각선 방향 체크
flag_b = [False] * (n*2-1)   # ↖  대각선 방향 체크


def put() -> None: # 함수 주석 -> / 반환하는 값이 None이라는 것을 의미
    # 각 열에 배치한 퀸의 위치를 출력
    for i in range(n):
        for j in range(n):
            print('■' if pos[i] == j else '□', end='')
        print()
    print()
    
count=0
def set(i: int) -> None:
    global count
    # i열에 퀸을 배치
    for j in range(n):
        if (    not flag[j]     # j행에 퀸이 없을 때
                and not flag_a[i+j]  # 대각선 방향으로 퀸이 없을 때
                and not flag_b[i-j+n-1]):  # 대각선 방향으로 퀸이 없을 때
            pos[i] = j  # 퀸을 j행에 배치
            if i == n-1:
                count+=1
                put()  # 모든 열에 퀸 배치를 끝내고 출력
            else:
                flag[j] = flag_a[i+j] = flag_b[i-j+n-1] = True
                set(i + 1)  # 다음 열에 퀸 배치
                flag[j] = flag_a[i+j] = flag_b[i-j+n-1] = False


set(0)  # 0 열에 퀸을 배치 시작
print(count)
```
