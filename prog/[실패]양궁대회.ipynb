```python
# https://think-tech.tistory.com/70
# itertools 에서 combinations with replacement를 이용한 방법 ( 중복조합 뽑기)

from itertools import combinations_with_replacement as cr
from collections import Counter

def solution(n, info):
    answer = []
    max_point=-1
    info=info[::-1]
    
    for i in cr(range(0,11),n):
        ry=0
        ap=0
        tmp=[0 for _ in range(11)]
        
        c = Counter(i)
        
        for j in range(0,11):
            if info[j]<c[j]:
                ry+=j
            elif info[j] != 0:
                ap+=j
                
            tmp[j]=c[j]
            
        if ry > ap:
            dif=ry-ap
            if max_point < dif:
                max_point=dif
                answer=tmp
                
    if answer :
        return answer[::-1]
    else :
        return [max_point]
```

```python
# combinations_with_replacement 를 통해 만들어진 i와 counter(i)를 확인
# i가 (0,0,3,0,10) 일경우 3번째 시도와 5번째 시도에 3점 10점을 냈음을 의미한다.
# 그때 counter(i)는 { 0:3,3:1,10:1 } 로 표현 0점을 3발 3점을 1발 10점을 1발 맞췄다.

for i in cr(range(0,11),5):
    c = Counter(i)
    print(i,c)
```

```python
# https://www.youtube.com/watch?v=dzncNbPMiB4
# 비트 쉬프트 연산 << 을 이용한 방법
# 이진수 코드를 만들어서 하는 방법이다
# 1<<1 => 2(10) / 1<<2 => 4(100) / 1<<3 => 8(1000)
# 이진수를 표현하는 방법 bin(i)

def solution(n, info):
    answer=[0]*11
    tmp=[0]*11
    max_diff=0
    
    # 과녁을 못맞추는 경우를 배제한다면 
    # 과녁에 맞췄다 못맞췄다라는 2진수의 형태로 표현할 수 있다.
    # 아래에 주어진 subset은 이진수 1(2)에서 1111111111(1023)까지를 표현
    for subset in range(1,1<<10):
        ry=0
        ap=0
        c=0
        
        for i in range(10):
            # 각 2진수의 자릿수의 값이 1인경우만 라이언이 이긴다고 체크
            if subset & (1<<i):
                ry += 10-i
                tmp[i]=info[i]+1
                c+= tmp[i]
            else:
                tmp[i]=0
                
                if info[i]:
                    ap+= 10 -i
                
        if c >n: continue
        
        tmp[10]=n-c
        
        if ry - ap == max_diff:
            for i in reversed(range(11)):
                if tmp[i]>answer[i]:
                    max_diff=ry-ap
                    answer = tmp[:]
                    break
                elif tmp[i]<answer[i]:
                    break
                
        elif ry-ap>max_diff:
            max_diff = ry-ap
            answer=tmp[:]
    
    if max_diff==0:
        answer=[-1]
        
```
