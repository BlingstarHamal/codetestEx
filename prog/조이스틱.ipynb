```python
# 12O 11P 10Q 9R 8S 7T 6U 5V 4W 3X 2Y 1Z 0A 1B 2C 3D 4E 5F 6G 7H 8I 9J 10K 11L 12M 13N
# 위아래 다음알파엣 이전알파벳(A이전은 Z)
# 커서 이동 처음에서 (왼쪽으로가면 맨뒤, 맨뒤에서 오른쪽으로 가면 맨앞)

# 현재 글자 기준으로 가까운방향 추적
# 각 알파벳에 번호를 메기고 |다음글자-현재글자|를 통해 최단 거리(상하) 분석
# 기본적으로 AAAA.....
```

```python
# 최단거거리 구하는 함수

def findShortestPath(name, now, next):
    right, left = max(next, now), min(next, now)
    rightDist = right - left
    leftDist = left + len(name) - right
    return min(rightDist, leftDist)


def solution(name):
    
    # A만 있는경우엔 그냥 리턴
    if set(name) == {'A'}:
        return 0

    alp={'A':0,'B':1,'C':2,'D':3,'E':4,'F':5,'G':6,'H':7,'I':8,'J':9,'K':10,'L':11,'M':12,'N':13,
        'O':12,'P':11,'Q':10,'R':9,'S':8,'T':7,'U':6,'V':5,'W':4,'X':3,'Y':2,'Z':1}
    
    answer = float('inf')
    count=0
    
    for i in range(len(name)):
        if name[i] == 'A':
            continue
        else :
            count+=alp[name[i]]
    
    # A를 제외하고 수를 남김
    toGoPlaces = [i for i in range(len(name)) if name[i] != "A" and i != 0]
    
    # 순열을 이용하여 모든 경우의수를 체크
    from itertools import permutations as p
    cases = p(toGoPlaces, len(toGoPlaces))
    
    for case in cases:
        now = 0
        result = 0
        for next in case:
            dist = findShortestPath(name, now, next)
            result += dist
            now = next
        answer = min(answer, result + count)
            
    return answer
```
